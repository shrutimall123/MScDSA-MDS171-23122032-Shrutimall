-> Inventory/shelf of items
-> Orders of customers

Create csv file which will store your inventory details ans order details

With the help of file handeling techniques in python, read the file and create an object for trinity store to populate the inventory items and orders into the trinity store.

To make sure that you have added all the items in your file, use a display method to show your inventory and order history.

```python
class SportMart:
    def __init__(self):
        self.inventory = []
        self.orders = []

    def add_item(self, item_name, price, quantity):
        self.inventory.append({"Item Name": item_name, "Price": price, "Quantity": quantity})

    def place_order(self, customer_name, item_name, quantity):
        item = next((item for item in self.inventory if item["Item Name"] == item_name), None)
        if item is not None and item["Quantity"] >= quantity:
            item["Quantity"] -= quantity
            self.orders.append({"Customer Name": customer_name, "Item Name": item_name, "Quantity": quantity, "Total Cost": item["Price"] * quantity})
            return True
        else:
            return False

    def display_inventory(self):
        print("SportMart Inventory:")
        for item in self.inventory:
            print(f"{item['Item Name']}: Price - ${item['Price']}, Quantity - {item['Quantity']}")

    def display_order_history(self):
        print("Order History:")
        for order in self.orders:
            print(f"Customer: {order['Customer Name']}, Item: {order['Item Name']}, Quantity: {order['Quantity']}, Total Cost: ${order['Total Cost']}")

    def save_to_csv(self):
        with open("inventory.csv", mode='w') as inventory_file:
            inventory_file.write("Item Name,Price,Quantity\n")
            for item in self.inventory:
                inventory_file.write(f"{item['Item Name']},{item['Price']},{item['Quantity']}\n")

        with open("order.csv", mode='w') as orders_file:
            orders_file.write("Customer Name,Item Name,Quantity,Total Cost\n")
            for order in self.orders:
                orders_file.write(f"{order['Customer Name']},{order['Item Name']},{order['Quantity']},{order['Total Cost']}\n")

    def load_from_csv(self):
        self.inventory = []
        self.orders = []

        with open("inventory.csv", mode='r') as inventory_file:
            inventory_file.readline()  # Skip the header
            for line in inventory_file:
                item_data = line.strip().split(",")
                self.add_item(item_data[0], float(item_data[1]), int(item_data[2]))

        with open("order.csv", mode='r') as orders_file:
            orders_file.readline()  # Skip the header
            for line in orders_file:
                order_data = line.strip().split(",")
                self.orders.append({
                    "Customer Name": order_data[0],
                    "Item Name": order_data[1],
                    "Quantity": int(order_data[2]),
                    "Total Cost": float(order_data[3])
                })

# Example Usage
trinity_store = SportMart()

# Add items to the inventory
trinity_store.add_item("Football", 20, 50)
trinity_store.add_item("Basketball", 15, 30)
trinity_store.add_item("Tennis Racquet", 40, 25)

# Place orders
trinity_store.place_order("Customer1", "Football", 2)
trinity_store.place_order("Customer2", "Basketball", 5)
trinity_store.place_order("Customer1", "Tennis Racquet", 3)

# Display the inventory and order history
trinity_store.display_inventory()
trinity_store.display_order_history()

# Save the data to CSV files
trinity_store.save_to_csv()

# Load data from CSV files
trinity_store.load_from_csv()

# Display the loaded data
trinity_store.display_inventory()
trinity_store.display_order_history()
```

Create a menu driven program that will
Create new orders and update the inventory accordingly
Export the final data to the file

```python
class SportMart:
    def __init__(self):
        self.inventory = {}
        self.orders = {}

    def create_inventory(self, ProductID, ProductName, Quantity, Price):
        self.inventory[ProductID] = {
            "ProductName": ProductName,
            "Quantity": Quantity,
            "Price": Price
        }

    def create_order(self, OrderId, ItemName, Quantity, Price):
        self.orders[OrderId] = {
            "ItemName": ItemName,
            "Quantity": Quantity,
            "Price": Price
        }

    def print_orders(self):
        for order_id, order_data in self.orders.items():
            print(f"Order ID: {order_id}, Item: {order_data['ItemName']}, Quantity: {order_data['Quantity']}")

    def print_inventory(self):
        for product_id, product_data in self.inventory.items():
            print(f"Product ID: {product_id}, Product Name: {product_data['ProductName']}, Quantity: {product_data['Quantity']}, Price: {product_data['Price']}")

trinity = SportMart()

# Load orders from the CSV file
with open("orders2.csv", "r") as orders_file:
    o_header = orders_file.readline()
    o_data = orders_file.readlines()
    for data in o_data:
        tmp = data.strip().split(',')
        trinity.create_order(tmp[0], tmp[1], int(tmp[2]), float(tmp[3]))

while True:
    print("Menu:")
    print("1. Add New Item to Inventory")
    print("2. Create a New Order")
    print("3. View Orders")
    print("4. View Inventory")
    print("5. Export Data to File")
    print("6. Exit")

    choice = int(input("Select your action:"))

    if choice == 1:
        ProductID = input("Enter Product ID: ")
        ProductName = input("Enter Product Name: ")
        Quantity = int(input("Enter Quantity: "))
        Price = float(input("Enter the Price of the Product: "))
        trinity.create_inventory(ProductID, ProductName, Quantity, Price)

    elif choice == 2:
        OrderId = input("Enter Order ID: ")
        ItemName = input("Enter Item Name: ")
        Quantity = int(input("Enter Quantity: "))
        Price = float(input("Enter Price: "))
        trinity.create_order(OrderId, ItemName, Quantity, Price)

    elif choice == 3:
        trinity.print_orders()

    elif choice == 4:
        trinity.print_inventory()

    elif choice == 5:
        # Export data to a file
        with open("exported_data.csv", "w") as export_file:
            export_file.write("Order ID,Item Name,Quantity,Price\n")
            for order_id, order_data in trinity.orders.items():
                export_file.write(f"{order_id},{order_data['ItemName']},{order_data['Quantity']},{order_data['Price']}\n")
            export_file.write("\nProduct ID,Product Name,Quantity,Price\n")
            for product_id, product_data in trinity.inventory.items():
                export_file.write(f"{product_id},{product_data['ProductName']},{product_data['Quantity']},{product_data['Price']}\n")
        print("Data has been exported to 'exported_data.csv'.")

    elif choice == 6:
        break

    else:
        print("Invalid choice. Please select a valid option.")
```
